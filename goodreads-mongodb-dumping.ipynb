```python
import pymongo
import pandas as pd
from pymongo import MongoClient
```

# Reading from MongoDB

```python
client = MongoClient("mongodb://localhost:27017/")
```

```python
db = client.goodreads
```

```python
gr_books_df = pd.DataFrame(list(db.books.find()))
```

```python
gr_reviews_df = pd.DataFrame(list(db.reviews.find({}, {'_id': 0, 'user_id': 0})))
```

```python
gr_authors_df = pd.DataFrame(list(db.authors.find({}, {'_id': 0})))
```

```python
gr_reviews_df = gr_reviews_df[gr_reviews_df.book_id.isin(gr_books_df.book_id)]
```

```python
gr_reviews_df.to_pickle('./gr_reviews.pkl')
```

```python
gr_books_df.to_pickle('./gr_books.pkl')
```

```python
gr_authors_df.to_pickle('./gr_authors.pkl')
```

# Reading from Pickle

```python
isbn_df = pd.read_pickle('../../isbn.pkl')
```

```python
gr_books_df = pd.read_pickle('./gr_books.pkl')
```

```python
gr_books_df.columns
```

```python
# Dropping books that have neither isbn nor isbn13
gr_books_df = gr_books_df[~((gr_books_df.isbn == '') & (gr_books_df.isbn13 == ''))]
```

```python
gr_books_df.shape
```

```python
# Dropping books that do not exist in our library system
gr_books_df = gr_books_df[(gr_books_df.isbn.isin(isbn_df.ISBN) | gr_books_df.isbn13.isin(isbn_df.ISBN))]
```

```python
gr_books_df.shape
```

```python
gr_isbns_df = gr_books_df[['book_id', 'isbn', 'isbn13']]
```
